## Project Overview

**Problem Statement**

Most personal finance tools stop at charts. They summarize spend but rarely explain why outcomes happen or what actions matter next.

**Why WalletMind exists**

WalletMind introduces an AI-first reasoning layer that converts statement data into explainable decisions, recommendations, and action plans.

**How AI agents solve the problem**

A Coordinator Agent orchestrates specialized agents (health, insights, budget, report, assistant) via deterministic Function Tools and shared services. This keeps outputs transparent, reproducible, and judge-friendly.

```mermaid
flowchart LR
    User --> Coordinator
    Coordinator --> Agents
    Agents --> FunctionTools
    FunctionTools --> Services
    Services --> Data
```

<a id='solution-overview'></a>
## Solution Overview

WalletMind uses a coordinator-led architecture to transform raw transaction data into explainable intelligence.

- **Coordinator**: selects execution strategy (single-agent or multi-agent) based on user intent.
- **Specialized Agents**: health, insights, budget, report, and assistant agents handle focused reasoning domains.
- **Function Tools**: deterministic boundaries between agents and WalletMind business services.
- **MCP Layer**: exposes WalletMind capabilities for MCP-compatible hosts.
- **Frontend Experience**: dashboard, agent playground, upload flow, and judge hub for transparent evaluation.
