## Executive Summary

WalletMind transforms raw bank statements into explainable, actionable financial intelligence.

- **Problem**: Traditional finance tools summarize history but rarely guide meaningful next actions.
- **Solution**: A coordinator-led multi-agent architecture built with Google ADK, connected to deterministic function tools and shared services.
- **Value**: Judges and users see transparent decision traces, per-agent outputs, and production-ready integration paths (REST + MCP).
- **Why agents**: Financial intelligence spans multiple specialized concerns; role-focused agents improve quality and explainability.

<a id='problem-statement'></a>
## Problem Statement

Traditional finance apps are often descriptive, not prescriptive.

### Pain Points
- Historical charts without clear next actions
- Opaque recommendation quality
- Limited trust due to weak explainability

### Why Multi-Agent for Financial Intelligence
- Distinct reasoning tasks (health, insights, budget, report, assistant) are better handled by specialized agents.
- Coordinator-based orchestration preserves coherence and traceability.
- Deterministic function tools keep business logic auditable and stable.

- `/mcp/tools`
- `/mcp/tools/{tool_name}/execute`

```python
def safe_get_json(url: str, timeout: int = 4):
    if requests is None:
        return False, {'error': 'requests package unavailable in runtime'}
    try:
        resp = requests.get(url, timeout=timeout)
        content_type = resp.headers.get('content-type', '')
        if 'application/json' in content_type:
            data = resp.json()
        else:
            data = {'status_code': resp.status_code}
        return True, {'status_code': resp.status_code, 'data': data}
    except Exception as exc:
        return False, {'error': str(exc)}

mcp_ok, mcp_result = safe_get_json(f'{MCP_BASE_URL}/mcp/metadata')
if mcp_ok:
    display(Markdown('### MCP Metadata Snapshot'))
    display(HTML(f"<pre style='background:#0f172a;color:#e2e8f0;padding:12px;border-radius:10px;overflow:auto'>{escape(json.dumps(mcp_result['data'], indent=2))}</pre>"))
else:
    render_callout('MCP Live Check', f"MCP metadata unavailable: {escape(mcp_result['error'])}", 'warning')
```

<a id='live-deployment-verification'></a>
## Live Deployment Verification

This automated check never hard-fails the notebook when services are temporarily offline.

```python
checks = [
    ('Frontend', FRONTEND_URL),
    ('Backend Root', REST_BASE_URL + '/'),
    ('REST Swagger', REST_SWAGGER_URL),
    ('MCP Root', MCP_BASE_URL + '/'),
    ('MCP Swagger', MCP_SWAGGER_URL),
    ('MCP Tools', MCP_BASE_URL + '/mcp/tools'),
]

def check_url(url: str) -> Dict[str, str]:
    if requests is None:
        return {'status': 'unavailable', 'http': '-', 'note': 'requests not installed'}
    try:
        response = requests.get(url, timeout=3)
        return {'status': 'reachable', 'http': str(response.status_code), 'note': 'ok'}
    except Exception as exc:
        return {'status': 'unavailable', 'http': '-', 'note': str(exc)[:120]}

status_rows: List[List[str]] = []
for name, url in checks:
    result = check_url(url)
    status_rows.append([name, url, result['status'], result['http'], result['note']])

render_table(headers=['Service', 'Endpoint', 'Status', 'HTTP', 'Note'], rows=status_rows)
```
